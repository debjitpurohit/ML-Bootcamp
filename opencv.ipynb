```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
#OPENCV module for image and video processing 
import cv2
```

```python
# frame=cv2.imread("strange.jpeg") #single face detection
frame=cv2.imread("group.jpg") # multiface detection
```

```python
plt.imshow(frame) #here (imread take the images in bgr format) its showing reverse of RGB (BGR)
```

```python
#converting BGR to RGB
#convert color
rgb=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

```python
# haar cascade is the training data for processing frontal faces ,eyes,anything...
# cascade frontalface download link ~ https://github.com/opencv/opencv/blob/4.x/data/haarcascades/haarcascade_frontalface_default.xml


#detecting frontal face 
classifier = cv2.CascadeClassifier("haarcascade_frontalface.xml")
faces=classifier.detectMultiScale(frame)
#faces having value like [[113  66  72  72]] beacuse this image have only single face
# for pointing out to the first face 
face=faces[1]
# #face having value like [113  66  72  72]
# #x ,y ,height ,width
x,y,w,h = face
# #and we have to cut the face using slicing from x to x+w and y to y+h from upper rgb image
finalface=rgb[y:y+h , x:x+w]
plt.imshow(finalface)
```

```python
# marking as rectangle on face from whole image
#here we have to pass imagename, point1(x,y) and point2(x+w,y+h) , color of reactangle in rgb ,thickness of reactangle
for face in faces:
    x,y,w,h = face
    out = cv2.rectangle(rgb,(x,y),(x+w,y+h),(255,0,0),6)
plt.imshow(out)
```
