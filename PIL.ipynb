```python
# for checking the pixels for strange.jpeg image
# (355,281,3) means let say 3 sqaure shaped layers(those are RED,GREEN,BLUE)  having 355 rows and 281 columns
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
pil = Image.open("strange.jpeg")
arr = np.array(pil)
arr.shape
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
pil2 = Image.open("rgb3.jpg")
arr = np.array(pil2)
arr.shape
plt.imshow(arr)
```

```python
pil2
```

```python
#if we want to extract only red color 
red = arr[:,:,0] #row colunmn layer
plt.imshow(red)
```

```python
#if we want to extract only green color 
green = arr[:,:,1] #row ,colunmn ,layer
plt.imshow(green)
```

```python
#if we want to extract only blue color 
blue = arr[:,:,2] #row ,colunmn ,layer
plt.imshow(blue)
```

```python
# cutting all the layer except red
work = arr.copy()
work[:,:,[1,2]]=200
plt.imshow(work)
```

```python
# generating black and white photo
# witout cmap its showing multi color gray scale
# axis 2 because we r taking mean of 3 layers
gray=arr.mean(axis=2)
plt.imshow(gray,cmap="gray")
```
