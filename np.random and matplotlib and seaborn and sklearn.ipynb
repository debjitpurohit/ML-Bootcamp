```python
#creating a circle
#as we know x=rcos0 and y =rsin0
r=10
theta=np.linspace(0,2*np.pi,200)
x=r*np.cos(theta)
y=r*np.sin(theta)
plt.figure(figsize=(10,10))
plt.plot(x,y)
```
